```python
import pandas as pd
```

```python
filename = 'psw18vwall.xls'
sheetname = 'Data 1'
```

```python
! pip install xlrd
```

```python
df = pd.read_excel(filename, sheet_name=sheetname)
df
```

```python
east_cost_oil_df = df[['Back to Contents', 'Unnamed: 2']].iloc[2:].copy()
east_cost_oil_df
```

```python
east_cost_oil_df.columns = ['date', 'retail_price']
east_cost_oil_df
# retail_price is retail price of oil in Dollars per Gallon at the east coast of US
```

```python
east_cost_oil_df.info()
```

```python
years = range(2019, 2024)
list(years)
```

```python
east_cost_oil_df['date'] = pd.to_datetime(east_cost_oil_df['date'])
east_cost_oil_df['retail_price'] = east_cost_oil_df['retail_price'].astype(float)

east_cost_oil_df.info()
```

```python
east_cost_oil_df
```

```python
east_cost_oil_df_2019_2023 = east_cost_oil_df[east_cost_oil_df['date'].dt.year.isin(years)]
east_cost_oil_df[east_cost_oil_df['date'].dt.year.isin(years)].index
```

```python
indexes = range(1295, 1557) # add one sample margin
print(list(indexes)[:5])
print(list(indexes)[-5:])
```

```python
east_cost_oil_df.loc[indexes]
```

```python
east_cost_oil_df_2019_2023 = east_cost_oil_df.loc[indexes]
east_cost_oil_df_2019_2023 = east_cost_oil_df_2019_2023.reset_index(drop=True)
east_cost_oil_df_2019_2023
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))

plt.plot(east_cost_oil_df_2019_2023['date'], east_cost_oil_df_2019_2023['retail_price'])

plt.xlabel('Date')
plt.ylabel('Price ($/gal)')
plt.title('Retail oli prices, USA, East Coast, Dollars per Gallon')

plt.show()
```

```python
east_cost_oil_df_2019_2023.to_csv('dataset_nyc_taxi/oil_prices.csv', index=False)
```
